### **Statistik Deskriptif**
* Jumlah anime dengan rating tinggi: 476
* Rata-rata rating: 8.43
* Rating tertinggi: 9.1, terendah: 8.1
* Rata-rata jumlah episode: 32 episode
* Rata-rata jumlah tema per anime: 2 tema

```python
df.rename(columns=lambda x: x.strip(), inplace=True)
high_rating_threshold = 8.0

high_rating_anime = df[df["Rating"] >= high_rating_threshold]

high_rating_stats = {
    "Jumlah Anime Rating Tinggi": high_rating_anime.shape[0],
    "Rata-rata Rating": high_rating_anime["Rating"].mean(),
    "Min Rating": high_rating_anime["Rating"].min(),
    "Max Rating": high_rating_anime["Rating"].max(),
    "Rata-rata Jumlah Episode": pd.to_numeric(high_rating_anime["Number_of_Episodes"], errors="coerce").mean(),
    "Rata-rata Jumlah Tema per Anime": high_rating_anime[[c for c in ["Theme1", "Theme2"] if c in high_rating_anime.columns]].count(axis=1).mean(),
}

high_rating_stats
```

### analisis Korelasi antara rating dan tema
*  Tidak ada hubungan langsung yang kuat antara satu tema dengan rating tinggi
* Korelasi antara tema dan rating tidak linier, tetapi lebih dipengaruhi oleh bagaimana cerita dan karakter dieksekusi

```python
theme1_ratings = high_rating_anime.groupby("Theme1")["Rating"].mean()
theme2_ratings = high_rating_anime.groupby("Theme2")["Rating"].mean()

theme_ratings = pd.concat([theme1_ratings, theme2_ratings]).groupby(level=0).mean()

top_themes_by_rating = theme_ratings.sort_values(ascending=False).head(10)

top_themes_by_rating
```

### **Penanganan**

* Menghapus data duplikat
* Menangani nilai hilang pada rating

```python
missing_percent = df.isnull().sum() / len(df) * 100
print(missing_percent)
```

```python
df_cleaned = df.drop_duplicates()

df_cleaned["Rating"].fillna(df_cleaned["Rating"].median(), inplace=True)
df_cleaned["Number_of_Episodes"].fillna(df_cleaned["Number_of_Episodes"].median(), inplace=True)

missing_values_after = df_cleaned.isnull().sum()
duplicate_rows_after = df_cleaned.duplicated().sum()

missing_values_after, duplicate_rows_after
```

* menangani nilai tidak konsisten

```python
def clean_theme(theme):
    if isinstance(theme, str):
        return theme.strip()
    return theme

df_cleaned["Theme1"] = df_cleaned["Theme1"].apply(clean_theme)
df_cleaned["Theme2"] = df_cleaned["Theme2"].apply(clean_theme)

df_cleaned = df_cleaned[~df_cleaned["Theme1"].str.match(r'^\d+$', na=False)]
df_cleaned = df_cleaned[~df_cleaned["Theme2"].str.match(r'^\d+$', na=False)]

cleaned_unique_themes1 = df_cleaned["Theme1"].unique()
cleaned_unique_themes2 = df_cleaned["Theme2"].unique()

cleaned_all_unique_themes = set(cleaned_unique_themes1).union(set(cleaned_unique_themes2))
sorted(cleaned_all_unique_themes)
```

### **Construct Data**

menambahkan kolom baru yang relevan untuk analisis tema vs rating. yaitu:
  * Jumlah tema per anime (menghitung apakah anime memiliki 1 atau 2 tema).
  * Kategori rating (misalnya: "Rendah", "Sedang", "Tinggi" berdasarkan rentang nilai rating).

```python
df_cleaned["Total_Themes"] = df_cleaned[["Theme1", "Theme2"]].notnull().sum(axis=1)

def categorize_rating(rating):
    if rating >= 8.0:
        return "Tinggi"
    elif rating >= 6.0:
        return "Sedang"
    else:
        return "Rendah"

df_cleaned["Rating_Category"] = df_cleaned["Rating"].apply(categorize_rating)

df_cleaned.head()
```
