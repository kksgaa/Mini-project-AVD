### **Penanganan**

```python
df_cleaned = df.drop_duplicates()

df_cleaned["Rating"].fillna(df_cleaned["Rating"].median(), inplace=True)
df_cleaned["Number_of_Episodes"].fillna(df_cleaned["Number_of_Episodes"].median(), inplace=True)

missing_values_after = df_cleaned.isnull().sum()
duplicate_rows_after = df_cleaned.duplicated().sum()

missing_values_after, duplicate_rows_after
```

* menangani nilai tidak konsisten

```python
def clean_theme(theme):
    if isinstance(theme, str):
        return theme.strip()
    return theme

df_cleaned["Theme1"] = df_cleaned["Theme1"].apply(clean_theme)
df_cleaned["Theme2"] = df_cleaned["Theme2"].apply(clean_theme)

df_cleaned = df_cleaned[~df_cleaned["Theme1"].str.match(r'^\d+$', na=False)]
df_cleaned = df_cleaned[~df_cleaned["Theme2"].str.match(r'^\d+$', na=False)]

cleaned_unique_themes1 = df_cleaned["Theme1"].unique()
cleaned_unique_themes2 = df_cleaned["Theme2"].unique()

cleaned_all_unique_themes = set(cleaned_unique_themes1).union(set(cleaned_unique_themes2))
sorted(cleaned_all_unique_themes)
```

### **Construct Data**

menambahkan kolom baru yang relevan untuk analisis tema vs rating. yaitu:
  * Jumlah tema per anime (menghitung apakah anime memiliki 1 atau 2 tema).
  * Kategori rating (misalnya: "Rendah", "Sedang", "Tinggi" berdasarkan rentang nilai rating).

```python
df_cleaned["Total_Themes"] = df_cleaned[["Theme1", "Theme2"]].notnull().sum(axis=1)

def categorize_rating(rating):
    if rating >= 8.0:
        return "Tinggi"
    elif rating >= 6.0:
        return "Sedang"
    else:
        return "Rendah"

df_cleaned["Rating_Category"] = df_cleaned["Rating"].apply(categorize_rating)

df_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## VISUALISASI DATA

### distribusi rating anime

Visualisasi distribusi rating menunjukkan bahwa sebagian besar anime memiliki rating antara 6 dan 8, dengan sedikit yang memiliki rating ekstrem (sangat rendah atau sangat tinggi).

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribusi Rating Anime')
plt.xlabel('Rating')
plt.ylabel('Jumlah Anime')
plt.show()
```

### Hitung kemunculan tema

Identifikasi tema yang paling sering muncul di anime dengan rating tinggi

```python
# Filter anime dengan rating tinggi (≥8)
high_rating_anime = df[df["Rating"] >= 8]

# Gabungkan Theme1 dan Theme2 dalam satu kolom, lalu hitung jumlah kemunculan setiap tema
theme_counts = pd.concat([high_rating_anime["Theme1"], high_rating_anime["Theme2"]]).value_counts()

# Visualisasi top 10 tema dengan jumlah anime terbanyak dalam kategori rating tinggi
plt.figure(figsize=(12, 6))
sns.barplot(x=theme_counts.values[:10], y=theme_counts.index[:10], palette="viridis")

plt.title("Top 10 Tema Anime dengan Rating Tinggi (≥8)")
plt.xlabel("Jumlah Anime")
plt.ylabel("Tema Anime")
plt.show()
```

### Menganalisis Tema dengan Rating Rata-rata Tertinggi

Selain menghitung jumlah kemunculan tema, kita juga ingin tahu tema mana yang memiliki rating rata-rata tertinggi. Ini penting karena tema yang sering muncul belum tentu memiliki rating yang sangat tinggi.

```python
# Gabungkan Theme1 dan Theme2 dalam satu kolom
melted_themes = pd.melt(high_rating_anime, id_vars=["Rating"], value_vars=["Theme1", "Theme2"], var_name="Theme_Type", value_name="Theme")
melted_themes = melted_themes.dropna()  # Hapus nilai NaN

# Hitung rata-rata rating per tema
theme_avg_rating = melted_themes.groupby("Theme")["Rating"].mean().sort_values(ascending=False).head(10)

# Visualisasi top 10 tema dengan rating tertinggi
plt.figure(figsize=(12, 6))
sns.barplot(x=theme_avg_rating.values, y=theme_avg_rating.index, palette="magma")

plt.title("Top 10 Tema dengan Rating Rata-rata Tertinggi (≥8)")
plt.xlabel("Rating Rata-rata")
plt.ylabel("Tema Anime")
plt.show()
```

### Analisis Korelasi Rating dengan tema

```python
# Menggabungkan semua tema dalam satu kolom untuk analisis yang lebih akurat
melted_themes = pd.melt(df, id_vars=["Rating"], value_vars=["Theme1", "Theme2"], var_name="Theme_Type", value_name="Theme")
melted_themes = melted_themes.dropna()  # Hapus nilai NaN

# Hitung rata-rata rating per tema
theme_avg_rating = melted_themes.groupby("Theme")["Rating"].mean().sort_values(ascending=False)

# Visualisasi rating rata-rata per tema
plt.figure(figsize=(12, 6))
sns.barplot(x=theme_avg_rating.index[:10], y=theme_avg_rating.values[:10], palette="coolwarm")

plt.xlabel("Tema Anime")
plt.ylabel("Rata-rata Rating")
plt.title("Rata-rata Rating untuk Setiap Tema Anime")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
